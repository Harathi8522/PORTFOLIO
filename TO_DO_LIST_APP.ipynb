```python
# Simple To-Do List App

tasks = []

def display_tasks():
    if not tasks:
        print("\nNo tasks yet!\n")
    else:
        print("\nYour Tasks:")
        for i, task in enumerate(tasks, 1):
            status = "[x]" if task["done"] else "[ ]"
            print(f"{i}. {status} {task['text']} (Priority: {task['priority']}, Due: {task['due']})")
        print()

while True:
    print("----- TO-DO LIST APP -----")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Mark Task as Completed")
    print("4. Delete Task")
    print("5. Unmark Task as Incomplete")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")

    if choice == "1":
        task_text = input("Enter task: ")
        priority = input("Enter priority (Low, Medium, High): ")
        due_date = input("Enter due date (e.g. 2025-06-15): ")
        tasks.append({
            "text": task_text,
            "done": False,
            "priority": priority,
            "due": due_date
        })
        print("Task added!\n")

    elif choice == "2":
        display_tasks()

    elif choice == "3":
        display_tasks()
        if tasks:
            task_num = input("Enter task number to mark as completed: ")
            if task_num.isdigit():
                index = int(task_num) - 1
                if 0 <= index < len(tasks):
                    tasks[index]["done"] = True
                    print("Task marked as completed!\n")
                else:
                    print("Invalid task number.\n")
            else:
                print("Please enter a number.\n")

    elif choice == "4":
        display_tasks()
        if tasks:
            task_num = input("Enter task number to delete: ")
            if task_num.isdigit():
                index = int(task_num) - 1
                if 0 <= index < len(tasks):
                    removed = tasks.pop(index)
                    print(f"Task '{removed['text']}' deleted!\n")
                else:
                    print("Invalid task number.\n")
            else:
                print("Please enter a number.\n")

    elif choice == "5":
        display_tasks()
        if tasks:
            task_num = input("Enter task number to mark as incomplete: ")
            if task_num.isdigit():
                index = int(task_num) - 1
                if 0 <= index < len(tasks):
                    tasks[index]["done"] = False
                    print("Task marked as incomplete.\n")
                else:
                    print("Invalid task number.\n")
            else:
                print("Please enter a number.\n")

    elif choice == "6":
        print("Goodbye! ðŸ‘‹")
        break

    else:
        print("Invalid choice. Please enter a number between 1 and 6.\n")
```
